```python
import os
import numpy as np
import librosa
import soundfile as sf
import tensorflow as tf
from tensorflow.keras import layers, models
from tqdm import tqdm
import IPython.display as ipd
import matplotlib.pyplot as plt
import scipy.signal as signal

def build_custom_model(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), padding='same', input_shape=input_shape))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(32, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Dropout(0.25))
    model.add(layers.Conv2D(64, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(64, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Dropout(0.25))
    model.add(layers.Conv2D(128, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(128, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Dropout(0.25))
    model.add(layers.UpSampling2D(size=(2, 2)))
    model.add(layers.Conv2D(128, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.UpSampling2D(size=(2, 2)))
    model.add(layers.Conv2D(64, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.UpSampling2D(size=(2, 2)))
    model.add(layers.Conv2D(32, (3, 3), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(2, (3, 3), activation='sigmoid', padding='same'))
    model.add(layers.ZeroPadding2D(padding=((0, 0), (0, 0))))
    model.compile(optimizer='adam', loss='mse')
    return model

input_shape = (1025, 216, 2)
custom_model = build_custom_model(input_shape)

def preprocess_track(track, n_fft=2048, hop_length=512):
    S = librosa.stft(track, n_fft=n_fft, hop_length=hop_length)
    S_db = librosa.amplitude_to_db(np.abs(S), ref=np.max)
    return S_db

def pad_or_truncate(spectrogram, freq_bins=1025, length=216):
    if spectrogram.shape[0] > freq_bins:
        spectrogram = spectrogram[:freq_bins, :]
    elif spectrogram.shape[0] < freq_bins:
        pad_height = freq_bins - spectrogram.shape[0]
        spectrogram = np.pad(spectrogram, ((0, pad_height), (0, 0)), mode='constant')
    if spectrogram.shape[1] > length:
        spectrogram = spectrogram[:, :length]
    elif spectrogram.shape[1] < length:
        pad_width = length - spectrogram.shape[1]
        spectrogram = np.pad(spectrogram, ((0, 0), (0, pad_width)), mode='constant')
    return spectrogram

def process_in_segments(model, mixture, segment_length=216, n_fft=2048, hop_length=512):
    mixture_spec_left = preprocess_track(mixture[0], n_fft=n_fft, hop_length=hop_length)
    mixture_spec_right = preprocess_track(mixture[1], n_fft=n_fft, hop_length=hop_length)
    num_segments = int(np.ceil(mixture_spec_left.shape[1] / segment_length))
    predicted_audio_segments = []
    for i in range(num_segments):
        start = i * segment_length
        end = (i + 1) * segment_length
        mixture_spec_left_segment = mixture_spec_left[:, start:end]
        mixture_spec_right_segment = mixture_spec_right[:, start:end]
        mixture_spec_left_segment = pad_or_truncate(mixture_spec_left_segment)
        mixture_spec_right_segment = pad_or_truncate(mixture_spec_right_segment)
        mixture_spec_segment = np.stack((mixture_spec_left_segment, mixture_spec_right_segment), axis=-1)
        mixture_spec_segment = np.expand_dims(mixture_spec_segment, axis=0)
        predicted_spec_segment = model.predict(mixture_spec_segment)
        predicted_left = librosa.istft(librosa.db_to_amplitude(predicted_spec_segment[0, :, :, 0]), hop_length=hop_length)
        predicted_right = librosa.istft(librosa.db_to_amplitude(predicted_spec_segment[0, :, :, 1]), hop_length=hop_length)
        predicted_audio_segment = np.vstack((predicted_left, predicted_right)).T
        predicted_audio_segments.append(predicted_audio_segment)
    predicted_audio = np.concatenate(predicted_audio_segments, axis=0)
    return predicted_audio

local_audio_path = 'vincent.wav'
mixture, sr = librosa.load(local_audio_path, sr=44100, mono=False)

isolated_vocals = process_in_segments(custom_model, mixture)

def simple_noise_reduction(audio, sr):
    sos = signal.butter(10, 1000, 'hp', fs=sr, output='sos')
    reduced_noise = signal.sosfilt(sos, audio)
    return reduced_noise

def apply_noise_reduction(isolated_vocals, sr, chunk_size=10):
    num_chunks = len(isolated_vocals) // int(chunk_size * sr)
    processed_chunks = []
    with tqdm(total=num_chunks, desc="Noise Reduction Progress") as pbar:
        for i in range(num_chunks):
            start = int(i * chunk_size * sr)
            end = int((i + 1) * chunk_size * sr)
            chunk = isolated_vocals[start:end]
            reduced_chunk = simple_noise_reduction(chunk, sr)
            processed_chunks.append(reduced_chunk)
            pbar.update(1)
    if len(isolated_vocals) % int(chunk_size * sr) != 0:
        start = num_chunks * int(chunk_size * sr)
        chunk = isolated_vocals[start:]
        reduced_chunk = simple_noise_reduction(chunk, sr)
        processed_chunks.append(reduced_chunk)
    concatenated_chunks = np.concatenate(processed_chunks)
    return concatenated_chunks

reduced_noise_vocals = apply_noise_reduction(isolated_vocals, sr, chunk_size=10)

max_val = np.max(np.abs(reduced_noise_vocals))
if max_val > 0:
    reduced_noise_vocals = reduced_noise_vocals / max_val

output_filename = 'isolated_vocals_reduced_VINCENT.wav'
sf.write(output_filename, reduced_noise_vocals, sr)

saved_audio, saved_sr = librosa.load(output_filename, sr=None)

plt.figure(figsize=(10, 4))
plt.plot(saved_audio)
plt.title("Waveform of the saved audio")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.show()

ipd.display(ipd.Audio(output_filename))
```
